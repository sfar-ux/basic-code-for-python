```python
a=[10,20,30,40,50]
#access the elements
print(a)
print(a[3])
print(a[-1])
#add of elements
a.append("world")
print(a)
a.insert(5,60)
print(a)
a.extend([70,80,90])
print(a)
# updation of elements
a[0]=1
print(a)
# del of elements
a.remove(1)
print(a)
a.pop(0)
print(a)
a.clear()
print(a)
del(a)
```

```python
#concatination
list1=[10,20,30,40]
list2=[50,60,70,80]
list3=list1+ list2
print(list3)
```

```python
#repetation
list1=[10,20]
list2=2*list1
print(list2)
```

```python
#slicing
a=[10,20,30,40,50,'hello']
print(a[0:6:2])
print(a[3:6])
```

```python
#sort the elements
a=[10,35,5,7,90,100]
b=sorted(a)
print(b)
#reverse the elements
a.reverse()
print(a)
#count
c=a.count(10)
print(c)
# length
length=len(a)
print(length)
#type
print(type(a))
```
