```python

# Create a dictionary and apply the following Methods: 1) Print the dictionary items 2) access items 3) use get() 4)change values 5) use len()
thisdict = {
  'brand': 'Ford',
  'model': 'Mustang',
  'year': 1964
}
# to print a dict
print(thisdict)
#To access the elements
print(thisdict["brand"])
#accessing the elements by using get()
print(thisdict.get("model"))
#update the data
thisdict["year"]="1974"
print(thisdict)
#using len() to find the no of data stored in dict
print(len(thisdict))
```

```python
#Write a program to find the factorial of a number using functions.

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result


num = int(input("Enter a number to find its factorial: "))


print("The factorial of ", num , "is", factorial(num))

```
