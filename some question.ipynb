```python
# 10. Factorial using recursion
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
n = int(input("Enter a number: "))
print("Factorial:", factorial(n))
```

```python
def count_vowels(input_string):
    # Define a set of vowels
    vowels = "aeiouAEIOU"
    count = 0

    # Iterate through each character in the string
    for char in input_string:
        if char in vowels:
            count += 1

    return count

# Get user input
user_input = input("Enter a string: ")
count = count_vowels(user_input)

print(f"The number of vowels in the given string is: {count}")
```

```python
# 11 Get the number of terms from the user
n = int(input("Enter the number of terms in the Fibonacci series: "))

# Initialize the first two terms
a, b = 0, 1
fibonacci_series = []

# Generate the Fibonacci series
for _ in range(n):
    fibonacci_series.append(a)
    a, b = b, a + b

# Print the Fibonacci series
print("Fibonacci series up to", n, "terms:", fibonacci_series)
```

```python
# 12
# Get the number from the user
num = int(input("Enter a number: "))

# Assume the number is prime
is_prime = True

# Check for factors
if num < 2:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

# Print the result
if is_prime:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
#13
# Get the range from the user
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

# List to hold prime numbers
prime_numbers = []

# Check each number in the range
for num in range(start, end + 1):
    if num < 2:
        continue
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(num)

# Print the prime numbers
print("Prime numbers in the range", start, "to", end, "are:", prime_numbers)
```
