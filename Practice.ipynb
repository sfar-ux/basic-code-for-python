```python
#1) Write a program to calculate roots of a polynomial equation.
# 15. Calculate roots of a polynomial equation
import cmath
def quadratic_roots(a, b, c):
    d = (b**2) - (4*a*c)
    root1 = (-b + cmath.sqrt(d)) / (2*a)
    root2 = (-b - cmath.sqrt(d)) / (2*a)
    return root1, root2

print("Enter coefficients a, b, and c for ax^2 + bx + c = 0:")
a, b, c = map(float, input().split())
roots = quadratic_roots(a, b, c)
print("Roots of the equation are:", roots)
```

```python
#3
# Program to check if a number is an Armstrong number

# Input from user
num = int(input("Enter a number: "))
num_str = str(num)
num_digits = len(num_str)
sum_of_cubes = sum(int(digit) ** num_digits for digit in num_str)

# Check and print result
if sum_of_cubes == num:
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```

```python
#4
# Illustrating tuples and operations on a tuple

# Creating a tuple
my_tuple = (10, 20, 30, 40, 50)
print("Original Tuple:", my_tuple)

# Accessing elements
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])

# Slicing a tuple
print("Slice from index 1 to 3:", my_tuple[1:4])

# Concatenation of tuples
tuple2 = (60, 70)
new_tuple = my_tuple + tuple2
print("Concatenated Tuple:", new_tuple)

# Repeating elements in a tuple
repeated_tuple = my_tuple * 2
print("Repeated Tuple:", repeated_tuple)

# Finding length of tuple
print("Length of Tuple:", len(my_tuple))

# Checking membership
print("Is 30 in tuple?", 30 in my_tuple)

# Iterating through tuple
print("Tuple elements:")
for item in my_tuple:
    print(item)

# Finding index of an element
print("Index of 40:", my_tuple.index(40))

# Counting occurrences of an element
tuple3 = (10, 20, 10, 30, 10)
print("Count of 10:", tuple3.count(10))
```

5.Reverse Indexing in Tuples in Python
In Python, tuples support reverse (negative) indexing, allowing access to elements from the end of the tuple.

```python
#5
# Define a tuple
tup = ('a', 'b', 'c', 'd', 'e')

# Accessing elements using negative indexing
print( tup[-1])     # 'e'
print( tup[-2])  # 'd'
print(tup[-3])   # 'c'

# Using reverse indexing in slicing

print(tup[::-1])  # ('e', 'd', 'c', 'b', 'a')
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

def calculator():
    print("Select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

    # Get user input for operation
    choice = input("Enter choice (1/2/3/4): ")

    # Get user input for numbers
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '1':
        print(f"{num1} + {num2} = {add(num1, num2)}")
    elif choice == '2':
        print(f"{num1} - {num2} = {subtract(num1, num2)}")
    elif choice == '3':
        print(f"{num1} * {num2} = {multiply(num1, num2)}")
    elif choice == '4':
        print(f"{num1} / {num2} = {divide(num1, num2)}")
    else:
        print("Invalid input")

# Run the calculator
calculator()
```

```python
#10
def is_prime(n):
    if n < 2:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Taking user input
num = int(input("Enter a number: "))

# Checking and printing the result
if is_prime(num):
    print(f"{num} is a Prime Number")
else:
    print(f"{num} is Not a Prime Number")
```
